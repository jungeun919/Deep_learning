```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import tensorflow as tf

# 실행할 때마다 같은 결과를 출력하기 위해 설정하는 부분
np.random.seed(3) # 한번 발생한 난수(무작위로 추출된 수)를 다시 표시할 때 사용
tf.random.set_seed(3) # 기준이 되는 seed 값을 정할 수 있게 해준다

Data_set = np.loadtxt("ThoraricSurgery.csv", delimiter=",") # delimiter은 csv가 ','로 구분됨

X = Data_set[:,0:17]
Y = Data_set[:,17]

model = Sequential() # 딥러닝의 구조를 한층한층 쉽게 쌓아올릴 수 있음
model.add(Dense(30, input_dim=17, activation='relu')) # Dense는 각 층의 특성을 옵션으로 함
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy']) # optimizer : 오차를 어떻게 줄여 나갈지 정하는 함수
model.fit(X, Y, epochs=100, batch_size=10)
```
